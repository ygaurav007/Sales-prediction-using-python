```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Datasets
df = pd.read_csv('advertising.csv')
# Returning top 5 rows of the dataset
df.head()
```

```python
#Returning a quick summary of the Dataset
df.info
```

```python
#Provides information about the number of rows and columns in the Dataset 
df.shape
```

```python
#Provides the description of the data in the Dataset
df.describe()
```

```python
#Visualizing pairwise relationships of sales and TV , Radio and newspaper respectively in a dataset. 
sns.pairplot(df, x_vars=['TV','Radio','Newspaper'], y_vars='Sales',kind='scatter')
plt.show()
```

```python
#Plotting a histogram for the 'TV' column of the DataFrame
df['TV'].plot.hist(bins=200,xlabel='TV' )
```

```python
#plotting a histogram for the 'Radio' column of the DataFrame
df['Radio'].plot.hist(bins=200,color='purple' ,xlabel="Radio")
```

```python
#Plotting a histogram for the 'Newspaper' column of the DataFrame
df['Newspaper'].plot.hist(bins=200,color='red' ,xlabel='Newspaper')
```

```python
#Generating a heatmap where colors represent the correlation strength between pairs of variables in the DataFrame
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#spliting arrays or matrices into random train and test subsets.
from sklearn.model_selection import train_test_split

#spliting the feature variable (X) and target variable (Y) into training and testing sets.
X_train, X_test, Y_train, Y_test = train_test_split(df[['TV']],df[['Sales']], test_size = 0.3, random_state = 0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
# Importing Logistic regression for binary classification
from sklearn.linear_model import LinearRegression

log = LinearRegression()
log.fit(X_train, Y_train)
```

```python
res=log.predict(X_test)
print(res)
```

```python
log.coef_
```

```python
0.05473199* 69.2 + 7.14382225
```

```python
plt.plot(res)
```

```python
plt.scatter(X_test,Y_test)
plt.plot(X_test,0.05473199* X_test + 7.14382225 , 'r')
plt.show()
```
